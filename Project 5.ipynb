# Visualizing The Gender Gap In College Degrees

In this project we will visualize the gender gaps in various College Degrees. The gender gap in the context of college degrees is that there is a statistical significant difference in the proportion of bachelor degree granted to women. The gap is reported often [in the news](https://www.google.com/search?hl=en&gl=us&tbm=nws&authuser=0&q=gender+gap+stem&oq=gender+gap+stem&gs_l=news) and [not everyone agrees](http://www.pbs.org/newshour/making-sense/truth-women-stem-careers/) that there is a gap. The aim of this project is to make a data-driven conclusion on the existence of the gender gap. 

We will be using the dataset containing the percentage of bachelor's degrees granted to women from 1970 to 2012. This dataset has been cleaned by Randal Olson who is a data scientist at University of Pennsylvania, and kindly made available [here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv). The dataset is broken up into 17 categories of degrees, with each column as a separate category. 

We will be using Pandas, matplotlib and various data visualization techniques which will take the center stage in assisting us towards making a data-driven conclusion. We will first import our dependencies for this project, which are pandas (imported as pd) and matplotlib.pyplot (imported as plt). Then we read in our dataset using Pandas and store this into the "woman_degrees" variables which we will be using throughout the project. Let's explore the dataset:

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
print(f"Dataset contains {women_degrees.shape[0]} rows and {women_degrees.shape[1]} columns")
women_degrees.head(5)
```

```python
women_degrees.describe(include='all')
```

As expected, there are 42 rows with rows being the data for each year and 18 columns with each column being the categories of degrees including the years column. The dataset itself is indeed very clean, where every category is a numerical field and there are no missing values. We are now ready to begin our analysis.

First we introduce the colour palette for this project below. We are colours that are friendly for people with colour blindness, so that everyone can enjoy the visualizations. We have chosen three colours within the **Color Blind 10** palette used by Tableau, which can be found [here](http://tableaufriction.blogspot.ro/2012/11/finally-you-can-use-tableau-data-colors.html).

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_gray = (171/255, 171/255, 171/255)
```

We will first explore the proportion of genders for the STEM degrees (**S**cience, **T**echnology, **E**ngineering and **M**aths). The STEM degrees are:

- Engineering
- Computer Science
- Psychology
- Biology
- Physical Sciences
- Math and Statistics

In the dataset, the proportions degrees granted to women are the actual percentage ranging from 0 to 100. So the proportion granted towards men must be "100 - proportion". We will plot the proportions granted towards women and the proportions granted towards men as line graphs both on one chart for each degree.

To maximise the impact of our visualizations, we will want to maximise the [**data-ink ratio** ](https://infovis-wiki.net/wiki/Data-Ink_Ratio) which is the proportion of ink to present the data. We will also take the appropriate measures to avoid misleading the audience. Here are the adjustments we will make:

- Remove the borders (i.e. spines) of each individual graph since it does not help in our visual analysis
- Setting a standard x-axis and y-axis scale to avoid misleading the audience
- Add in two sets of label to communicate the meaning of each line

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)

    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom = False, top = False, left = False, right = False)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
plt.show()
```

As we can see from just the STEM degrees, there is clearly a gender gap for Engineering, Computer Science and Psychology degrees. We can see for Psychology, the gap has widened over time and for Physical Sciences, the gap has closed over time. Computer Science and Biology degrees saw some fluctuation in the gender gap over time. The gender gap is smallest for Maths and Statistics. 

Next we will extend this analysis across all degrees.

## Comparing across all degrees

In this section, we group the degrees into STEM, liberal arts and other in the following way:

| STEM                | Liberal Arts                  | Other Categories      |
|---------------------|-------------------------------|-----------------------|
| Psychology          | Foreign Languages             | Health Professions    |
| Biology             | English                       | Public Administration |
| Math and Statistics | Communications and Journalism | Education             |
| Physical Sciences   | Art and Performance           | Agriculture           |
| Computer Science    | Social Sciences and History   | usiness               |
| Engineering         |                               | Architecture          |

We are also aiming to plot similar graphs to the ones in the previous section, except in the order in the table above. Rather than writing everything from scratch, we will modify our previous code to match this situation.

- First we place all our degree names into their respective lists, then append "" to the end of liberal arts so that every list has 6 entires
- Then we use the *zip* Python built in function to group all the $n^{th}$ entry together and then use the *list* method to turn the zip object back into a list of tuples
- Then use a double for loop to iterate over that new list, where in each iteration runs an identical code in the previous section

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
lib_arts_cats.append("")

all_cats = list(zip(stem_cats, lib_arts_cats, other_cats))
fig = plt.figure(figsize=(18, 18))

for y in range(6):
    for x in range(3):
        if x == 1 and y == 5:
            continue
        ax = fig.add_subplot(6,3,x+(3*y)+1)
        ax.plot(women_degrees['Year'], women_degrees[all_cats[y][x]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[all_cats[y][x]], c=cb_orange, label='Men', linewidth=3)
        
        for key in ax.spines.items():
            ax.spines[key[0]].set_visible(False)
            
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(all_cats[y][x])
        ax.tick_params(bottom = False, top = False, left = False, right = False)
        
        if x == 0 and y == 0:
            ax.text(2006, 12, 'Men')
            ax.text(2005, 84, 'Women')
        elif x == 1 and y == 0:
            ax.text(2006, 18, 'Men')
            ax.text(2005, 78, 'Women')    
        elif x == 2 and y == 0:
            ax.text(2006, 6, 'Men')
            ax.text(2005, 90, 'Women')
        elif x == 0 and y == 5:
            ax.text(2006, 87, 'Men')
            ax.text(2005, 6, 'Women')       
        elif x == 2 and y == 5:
            ax.text(2006, 63, 'Men')
            ax.text(2005, 33, 'Women')        
        
plt.show()
```

As we can see, the x-axis labels overlaps with the titles of the graph underneath, next we will clean up our visualization further.

## Hiding x-axis labels

To improve the aesthetics of our visualization, we will hide all x-axis labels in the column except for the bottom one. This way we can remove the overlaps while the audience would still know the scale by referring to that bottom most graph for each column.

```python
fig = plt.figure(figsize=(18, 18))

for y in range(6):
    for x in range(3):
        if x == 1 and y == 5:
            continue
        ax = fig.add_subplot(6,3,x+(3*y)+1)
        ax.plot(women_degrees['Year'], women_degrees[all_cats[y][x]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[all_cats[y][x]], c=cb_orange, label='Men', linewidth=3)
        
        for key in ax.spines.items():
            ax.spines[key[0]].set_visible(False)
            
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(all_cats[y][x])
        
        if ( y == 5 ) or ( y == 4 and x == 1 ):
            ax.tick_params(bottom = False, top = False, left = False, right = False)
        else:
            ax.tick_params(bottom = False, top = False, left = False, right = False, labelbottom = False)
        
        if x == 0 and y == 0:
            ax.text(2006, 12, 'Men')
            ax.text(2005, 84, 'Women')
        elif x == 1 and y == 0:
            ax.text(2006, 18, 'Men')
            ax.text(2005, 78, 'Women')    
        elif x == 2 and y == 0:
            ax.text(2006, 6, 'Men')
            ax.text(2005, 90, 'Women')
        elif x == 0 and y == 5:
            ax.text(2006, 87, 'Men')
            ax.text(2005, 6, 'Women')       
        elif x == 2 and y == 5:
            ax.text(2006, 63, 'Men')
            ax.text(2005, 33, 'Women')        
        
plt.show()
```

The graph looks much cleaner, however there are too many y-axis labels.

## Setting y-axis labels

To further clean up our visualization, we will only show the ending labels 0 and 100.

```python
fig = plt.figure(figsize=(18, 18))

for y in range(6):
    for x in range(3):
        if x == 1 and y == 5:
            continue
        ax = fig.add_subplot(6,3,x+(3*y)+1)
        ax.plot(women_degrees['Year'], women_degrees[all_cats[y][x]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[all_cats[y][x]], c=cb_orange, label='Men', linewidth=3)
        
        for key in ax.spines.items():
            ax.spines[key[0]].set_visible(False)
            
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(all_cats[y][x])
        ax.set_yticks([0,100])
        
        if ( y == 5 ) or ( y == 4 and x == 1 ):
            ax.tick_params(bottom = False, top = False, left = False, right = False)
        else:
            ax.tick_params(bottom = False, top = False, left = False, right = False, labelbottom = False)
        
        if x == 0 and y == 0:
            ax.text(2006, 12, 'Men')
            ax.text(2005, 84, 'Women')
        elif x == 1 and y == 0:
            ax.text(2006, 18, 'Men')
            ax.text(2005, 78, 'Women')    
        elif x == 2 and y == 0:
            ax.text(2006, 6, 'Men')
            ax.text(2005, 90, 'Women')
        elif x == 0 and y == 5:
            ax.text(2006, 87, 'Men')
            ax.text(2005, 6, 'Women')       
        elif x == 2 and y == 5:
            ax.text(2006, 63, 'Men')
            ax.text(2005, 33, 'Women')        
        
plt.show()
```

The visualization looks much cleaner, now we will add some final touches and we will be ready to export it.

## Adding a horizontal line

By removing y-axis labels as well as many of the other features, it has made it much more challenging to properly gauge gender gaps. To solve this, we will add a horizontal line across all line charts where the y-axis label 50 would have been. In order to not confuse the audience, we will also give that line a label.

```python
from matplotlib.patches import Rectangle
fig = plt.figure(figsize=(18, 18))

for y in range(6):
    for x in range(3):
        if x == 1 and y == 5:
            continue
        ax = fig.add_subplot(6,3,x+(3*y)+1)
        ax.plot(women_degrees['Year'], women_degrees[all_cats[y][x]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[all_cats[y][x]], c=cb_orange, label='Men', linewidth=3)
            
        for key in ax.spines.items():
            ax.spines[key[0]].set_visible(False)
            
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(all_cats[y][x])
        ax.set_yticks([0,50,100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        
        ax.add_patch(Rectangle((0,0),6,3, lw=1, fill=False))
        
        if ( y == 5 ) or ( y == 4 and x == 1 ):
            ax.tick_params(bottom = False, top = False, left = False, right = False)
        else:
            ax.tick_params(bottom = False, top = False, left = False, right = False, labelbottom = False)
        
        if x == 0 and y == 0:
            ax.text(2006, 12, 'Men')
            ax.text(2005, 84, 'Women')
        elif x == 1 and y == 0:
            ax.text(2006, 18, 'Men')
            ax.text(2005, 78, 'Women')    
        elif x == 2 and y == 0:
            ax.text(2006, 6, 'Men')
            ax.text(2005, 90, 'Women')
        elif x == 0 and y == 5:
            ax.text(2006, 87, 'Men')
            ax.text(2005, 6, 'Women')       
        elif x == 2 and y == 5:
            ax.text(2006, 63, 'Men')
            ax.text(2005, 33, 'Women')        

plt.show()
```
