# Visualizing Earnings Based On College Majors

The aim of this project is to explore the dataset visually using Pandas. Since our brains have evolved to develop power visual processing capabilities, using data visualization is much faster and easier to uncover patterns and insights behind our data compared to a table of numbers and figures. While matplotlib can create a much larger variety of charts, Pandas has many methods which more quickly and easily generate common plots from data. We will still utilise matplotlib to customise the plots, however in this project, Pandas will take the centre stage.

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

| Column Name     | Column Description                                                 |
|-----------------|--------------------------------------------------------------------|
| Rank            | Rank by median earnings (the dataset is   ordered by this column). |
| Major_code      | Major code.                                                        |
| Major           | Major description.                                                 |
| Major_category  | Category of major.                                                 |
| Total           | Total number of people with major.                                 |
| Sample_size     | Sample size (unweighted) of full-time.                             |
| Men             | Male graduates.                                                    |
| Women           | Female graduates.                                                  |
| ShareWomen      | Women as share of total.                                           |
| Employed        | Number employed.                                                   |
| Median          | Median salary of full-time, year                                   |
| Low_wage_jobs   | Number in low-wage service jobs.                                   |
| Full_time       | Number employed 35 hours or more.                                  |
| Part_time       | Number employed less than 35 hours.                                |

We can quickly answer questions such as using the appropriate visualisation, where:

- Using scatter plots: Do students in more popular majors make more money?
- Using histograms: How many majors are predominantly male? Predominantly female?
- Using bar plots: Which category of majors have the most students?

The first step in the Exploratory Data Visualization process is to import the pandas and matplotlib.pyplot. The line "%matplotlib inline" enables Jupyter and matplotlib to work together effectively, where Jupyter will automatically show the plot of the current cell once run. 

After importing the appropriate libraries, we will now read the data using Pandas. We will first explore the data before we visualize it.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv", encoding="UTF-8")
recent_grads
```

```python
recent_grads.describe()
```

One thing to note is that the rows seems to be ordered by Median salary. 

All the columns we would want to analyze we want to analyze are numeric, therefore we do not need to adjust the data before we begin our analysis. The only cleaning step we need to do is the remove the rows with null values since matplotlib expects columns of values we pass in to have matching legnths, and missing vlaues will cause matplotlib to throw errors. 

```python
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]

print(f"Number of Rows prior to cleaning: {raw_data_count}")
print(f"Number of Rows after to cleaning: {cleaned_data_count}")
```

As expected, there are 172 rows after cleaning, which is in line with the information in the general summary statistics of the numeric columns. 

Next, we will start to visualize the data in various plots.

## Pandas, Scatter Plots

We will first use the **Scatter Plot**. Scatter plots allows us to help us determine the correlation between the two columns. it also allows us to intuitively sense how spread out the data is, and compare easily. We will generate the following scatter plots:

- Median vs. Sample_size
- Unemployment_rate vs. Sample_size
- Median vs. Full_time
- Unemployment_rate vs. ShareWomen
- Median vs. Men
- Median vs. Women

Using these plots we aim to answer the following questions:

- Do students in more popular majors make more money?
- Do students that majored in subjects that were majority female make more money?
- Is there any link between the number of full-time employees and median salary?

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title("Median vs. Sample_size")
```

Although we are able to successfully generate a scatter plot, most of our data is centred near the bottom left of the plot. We will zoom into that area of the graph, which has the biggest impact on the correlation between the median wage and the popularity of the majors. However, this is at the cost of ignoring the outliers.

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title("Median vs. Sample_size (zoomed-in)")
ax.set_xlim(0,1500)
ax.set_ylim(20000,80000)
```

From this chart, there is a very weak negative correlation between the median wage and popularity of the major. A possible explanation to this is there is still a demand for jobs with unpopular majors. Due to the low supply of fresh graduates, employers will need to pay a higher salary to entice to potential employee for the position as well as to maintain existing employees so that they stay with the current employer. 

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title("Unemployment_rate vs. Sample_size")
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title("Median vs. Full_time")
```

Since this plot is crucial in helping us answer one of the questions we posed, we will zoom into the highly clustered area for a better view.

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title("Median vs. Full_time (zoomed-in)")
ax.set_xlim(0,50000)
ax.set_ylim(20000,70000)
```

We can see that this plot does exhibit a very weak negative correlation (at best) between Median salary and the number employed more than 35 hours. A possible explanation for this is that there needs to be a higher salary to meet the required demand when there is a shortage of potential employee with the required skillset.

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title("Unemployment_rate vs. ShareWomen")
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title("Median vs. Men")
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title("Median vs. Women")
```

We will plot another scatter plot (Median vs. ShareWomen) to answer the last question.

```python
ax = recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
ax.set_title("Median vs. ShareWomen")
```

It is evident that there is a negative correlation between the median salary and the proportion of women who graduated in the particular major. Therefore, students who majored in subjects with a higher proportion of women tends to make less money.

## Pandas, Histograms

Next we will visualize our data using **Histograms**, which is a visual representation of a frequency distribution. This allows us to see the distubution of our data.

Using Histograms, we will visually explore the distributions of the following columns:

- Sample_size
- Median
- Employed
- Full_time
- ShareWomen
- Unemployment_rate
- Men
- Women

```python
def hist_generator(series,b):
    ax = series.hist(bins=b)
    ax.set_xlabel(series.name)
    ax.set_ylabel("Count")
    ax.set_title(f"Count of {series.name} split across {b} bins")
    return ax

hist_generator(recent_grads["Sample_size"], 25)
```

Majority of majors tends to have a very small sample size.

```python
hist_generator(recent_grads["Median"], 25)
```

Majority of majors have median salaries at around $30,000 to $50,000.

```python
hist_generator(recent_grads["Employed"], 25)
```

Majority of majors tends to have less than 50,000 employees, which is far smaller than the maximum of over 300,000.

```python
hist_generator(recent_grads["Full_time"], 25)
```

Majority of the majors tends to have less than 40,000 full-time employees, which is far smaller than the maximum value.

```python
hist_generator(recent_grads["ShareWomen"], 25)
```

There is a slight negative skewness in the data, suggesting that most of the majors have a higher proportion of women who graduated. This in line with the statistic where there is a higher proportion of woman who graudate from college.

```python
hist_generator(recent_grads["Unemployment_rate"], 25)
```

This distribution has some resemblance to a normal distribution, with some outliers towards the tails.

```python
hist_generator(recent_grads["Men"], 25)
```

```python
hist_generator(recent_grads["Women"], 25)
```

The histogram of Men and Women follows a very similar pattern where there is a high majority in the smaller end of the values.

## Pandas, Scatter Matrix Plot

We will be using the **Scatter Matrix Plot** next, this is a very useful plot for the purposes of exploratory analysis as it combines both the scatter plot and histogram together. In this visualisation, the histograms makes up the diagonals and the scatter plots makes up the off-diagonals. We will first plot a $2 \times 2$ scatter plot, followed by a $3 \times 3$ one. While it is possible to increase the dimensions, the plots becomes increasingly unreadable and difficult to interpret beyond columns of 4 or more, therefore we will stop at 3 columns.

```python
from pandas.plotting import scatter_matrix

scatter_matrix(recent_grads[["Sample_size", "Median"]], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[["Sample_size", "Median", "Unemployment_rate"]], figsize=(15,15))
```

Although the scatter matrix provides a lot of information, we are not able to have much customisation over the plots, which make it difficult to highlight insights necessary to answer key questions. 

## Pandas, Bar Plots

The last chart we will use is the **Bar Chart**, which is helps us compare summary statistics across different categories in a visual manner, which aids in gaining a more intuitive understanding of the data.

To ensure the visualisation stays intuitive, we must make sure the charts use the a shared scale. It is important to note that the plots needs to be large enough so the information remains clear when compared side by side.

For our analysis, we will compare percentage of women and the employment rate for the top ten and bottom ten rows of our dataset. 

```python
fig, axs = plt.subplots(1,2,figsize=(12.8,4.8))

ax1 = recent_grads[:10].plot.bar(x="Major", y="ShareWomen", ax=axs[0])
ax1.set_ylim(0,1)
ax2 = recent_grads[-10:].plot.bar(x="Major", y="ShareWomen", ax=axs[1])
ax2.set_ylim(0,1)
```

The top ten rows contains majors which have a much lower proportion of woman compared to the bottom ten rows. Upon closer inspection, the top ten rows contains STEM subjects which are typically have low porportion of woman, while the bottom ten subjects are humanity type subjects which typically taken by woman.

```python
fig, axs = plt.subplots(1,2,figsize=(12.8,4.8))

ax1 = recent_grads[:10].plot.bar(x="Major", y="Unemployment_rate", ax=axs[0])
ax1.set_ylim(0,0.175)
ax2 = recent_grads[-10:].plot.bar(x="Major", y="Unemployment_rate", ax=axs[1])
ax2.set_ylim(0,0.175)
```

The employment rate tends to be lower for the top ten rows compared to the bottom ten rows.

From the scatter plots:

In the context of this analysis, many of the majors will have a small count since majority of the students will have enrolled in a few select majors.
